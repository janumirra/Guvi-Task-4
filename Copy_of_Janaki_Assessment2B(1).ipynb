###### merge those two csv files (after getting as dataframes, get them as a single dataframe)

###### consider if the codekata score exceeds 15000 points(present week) then make a csv on those observations as Exceeded expectations.csv


```python
df_merged.rename(columns = {'Previous Geekions':'Previous_Geekions', 'CodeKata Score':'CodeKata_Score'}, inplace = True)
#ck15k=codekata >=15k
ck15k=df_merged.query("CodeKata_Score>15000")

ck15k.to_csv("Exceeded expectations.csv")
ck15k
```

###### Average of previous week geekions vs this week geekions (i.e Previous Geekions vs CodeKata Score)

```python
avg_Pg=df_merged['Previous_Geekions'].describe()
avg_Cs=df_merged['CodeKata_Score'].describe()
p_vs_c=pd.DataFrame({"Previous_Geekoins":avg_Pg,
                    "Codekata_score":avg_Cs}
                    )
p_vs_c
```

```python
p_vs_c["Statistics"]=p_vs_c.index.values
p_vs_c
```

```python
fig, axes = plt.subplots(1, 2, figsize=(15, 5), sharey=True)
fig.suptitle('Previous Geekions vs CodeKata Score')

# Previous_Geekoins
fig1=sns.barplot(ax=axes[0], x=p_vs_c["Statistics"], y=p_vs_c['Previous_Geekoins'])
axes[0].set_title('Previous_Geekoins')
for j in fig1.containers:
  fig1.bar_label(j,label_type='edge')


# Codekata_score
fig2=sns.barplot(ax=axes[1], x=p_vs_c["Statistics"], y=p_vs_c['Codekata_score'])
axes[1].set_title('Codekata_score')
for i in fig2.containers:
  fig2.bar_label(i,label_type='edge')
plt.show()

```

###### #Average completion of python course or my_sql or python english or computational thinking

```python
df_merged=df_merged.fillna(value=0)
completion=df_merged["python"].mean()
completion
```

```python
df_merged=df_merged.fillna(value=0)
avg=df_merged.groupby(["Department"],as_index=False).mean()
print(avg)

fig=sns.catplot(
    data=avg,kind='bar',
    x='Department',
    y='python',
    
    )
# iterate through the axes
for ax in fig.axes.flat:

    # annotate
    ax.bar_label(ax.containers[0], label_type='edge')

    # pad the spacing between the number and the edge of the figure; should be in the loop, otherwise only the last subplot would be adjusted
    ax.margins(y=0.1)
plt.title('Average completion of Python', loc='center',color='r')
plt.xticks(fontsize=10,rotation = 45,color='black')
plt.show()
```

###### Department wise codekata performence (pie chart)

```python
import pandas as pd
data=df_merged.groupby(["Department"],as_index=False).mean()
print(data)

df_dep=pd.DataFrame(data[['Department',"CodeKata_Score"]])
print(df_dep)
labels=df_dep["Department"]
data1=df_dep['CodeKata_Score']

fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.pie(data1,labels=labels,autopct="%1.2f%%")
plt.title('Department wise codekata performence', loc='center',color='r')


plt.plot()
```

###### Department wise toppers (horizantal bar graph or any visual representations of your choice)

```python
dep_topper=df_merged.groupby(["Department"],as_index=False).max()
print(dep_topper)
dep_top_df=dep_topper[["Name","Department","CodeKata_Score"]]

print(dep_top_df)
dfff=pd.DataFrame(dep_top_df)
dfff
```

```python
plt.figure(figsize=(12, 5))
fig=sns.barplot(x='CodeKata_Score',y='Name',data=dfff,hue=dfff['Department'])
plt.title('Department wise Toppers', loc='center',color='r')
plt.legend(bbox_to_anchor=(1.01, 1), loc='upper left', borderaxespad=1)
for i in fig.containers:
  fig.bar_label(i,label_type='edge')
plt.show()
```
